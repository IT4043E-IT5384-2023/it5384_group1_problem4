```python
new_wallet_df = all_wallet_df[['_id', 'address','tokens','balanceInUSD','borrowInUSD','borrowTokens', 'dailyNumberOfTransactions','dailyAllTransactions', 'dailyTransactionAmounts', 'depositInUSD', 'type', 'numberOfLiquidation', 'totalValueOfLiquidation', '_id_contract', 'address_contract', 'name_contract', '_id_project','name_project', 'category_project', 'socialAccounts_project']]
new_wallet_df
```

```python
nan_count = new_wallet_df.isna().sum()

print(nan_count)
```

```python
import ast
def adding_values(x):
  if x is None or x == 'NaN':
    return 0
  elif isinstance(x, float) == False and isinstance(x, int) == False and x is not None:
    return sum(ast.literal_eval(x).values())
new_wallet_df['tokens'] = new_wallet_df['tokens'].apply(adding_values)
new_wallet_df['borrowTokens'] = new_wallet_df['borrowTokens'].apply(adding_values)
new_wallet_df['dailyNumberOfTransactions'] = new_wallet_df['dailyNumberOfTransactions'].apply(adding_values)
new_wallet_df['dailyAllTransactions'] = new_wallet_df['dailyAllTransactions'].apply(adding_values)
new_wallet_df['dailyTransactionAmounts'] = new_wallet_df['dailyTransactionAmounts'].apply(adding_values)
new_wallet_df
```

```python
new_wallet_df.to_csv('cleaned_data.csv')
```
