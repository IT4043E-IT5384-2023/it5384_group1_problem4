# Merge

```python
project_constract = projects.merge(smart_contracts, left_on = 'new_addresses_project', right_on="_id_contract",how = "inner")
project_constract
```

```python
transact_project = project_constract.merge(token_transfer, left_on="address_contract", right_on="contract_address_transact", how="inner")
transact_project
```

```python
wallet_project = wallets.merge(transact_project, left_on ="address", right_on="from_address_transact", how='left')
wallet_project["is_sender"] = wallet_project['from_address_transact'].notnull()
wallet_project
```

```python
wallet_project_1 = wallets.merge(transact_project, left_on ="address", right_on="to_address_transact", how='left',suffixes=('_receiver',''))
wallet_project_1["is_receiver"] = wallet_project_1['to_address_transact'].notnull()
wallet_project_1
```

```python
[(i, len(wallet_project[i].drop_duplicates())) for i in wallet_project.columns]
```

```python
cols = ["_id","_id_contract","transaction_hash_transact"]
len(wallet_project[cols].drop_duplicates())
```

```python
wallet_project_full = wallet_project.merge(wallet_project_1, on = ["_id","_id_contract","transaction_hash_transact"],how="outer")
```

```python
wallet_project_full
```

```python
# wallet_project_full["_id"].drop_duplicates()
wallet_project_full = wallet_project_full[(wallet_project_full["_id_project"].notnull())]
```

```python
wallet_project_full['type'] = 'single_chain'
```

```python
wallet_project_full
```

```python
multichain_wallet_project = multichain_wallets.merge(transact_project, left_on ="address", right_on="from_address_transact", how='left',suffixes=('_sender', ''))
multichain_wallet_project["is_sender"] = multichain_wallet_project['from_address_transact'].notnull()

multichain_wallet_project_1 = multichain_wallets.merge(transact_project, left_on ="address", right_on="to_address_transact", how='left',suffixes=('_receiver',''))
multichain_wallet_project_1["is_receiver"] = multichain_wallet_project_1['to_address_transact'].notnull()

multichain_wallet_project_full = multichain_wallet_project.merge(multichain_wallet_project_1[["_id","_id_contract","transaction_hash_transact","is_receiver"]], on = ["_id","_id_contract","transaction_hash_transact"],how="outer")
multichain_wallet_project_full = multichain_wallet_project_full[(multichain_wallet_project_full["_id_project"].notnull())]
```

```python
multichain_wallet_project_full['type'] = 'multichain_wallet'
multichain_wallet_project_full
```

```python
print(len(multichain_wallet_project_full["_id"].drop_duplicates()))
print(len(wallet_project_full["_id"].drop_duplicates()))
```

```python
all_wallet= pd.concat([multichain_wallet_project_full,wallet_project_full])
```
